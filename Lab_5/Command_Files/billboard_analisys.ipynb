```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("C:\\Users\\DELL\\Desktop\\Air_semestr_5\\AIBD\\Adam_Stasiak_AIBD\\Lab_5\\Analysis_Data\\Billboard_da.csv",encoding='unicode_escape')
data
```

```python
data_table = pd.read_excel("C:\\Users\\DELL\\Desktop\\Air_semestr_5\\AIBD\\Adam_Stasiak_AIBD\\Lab_5\\Analysis_Data\\billboard_table.xlsx")
data_table
```

```python
colum_names =[]
for row in data_table:
    colum_names.append(row)
print(colum_names)
```

```python
weeks = colum_names[7:]
highest_ranking  = []
mean_ranking = []
weeks_df = data_table[weeks]
# weeks_to_mean = weeks_df.replace(np.NaN,0)
# weeks_df= weeks_df.replace(np.NaN,np.inf)
mean = weeks_df.mean(axis=1,skipna=True)
mins = weeks_df.min(axis=1,skipna=True)

for i in range(len(mins)):
    highest_ranking.append(mins[i])
    mean_ranking.append(mean[i])

```

```python
columns_to_drop = colum_names[6:]
#print(columns_to_drop)
data_table= data_table.drop(columns=columns_to_drop)

```

Przetworzona tabela

```python
data_table.insert(6,"highest ranking",highest_ranking,allow_duplicates=True)
data_table.insert(7,"mean ranking",mean_ranking,allow_duplicates=True)
data_table
```

Przykładowe wykresy

Wykres najwyższego rankingu dla poszczególnych utworów

```python
data_table.plot(x='artist.inverted',y='highest ranking',kind='bar',figsize=(50,50),title='Najwyższy ranking dla danego utworu')
```

Średni ranking dla utworu

```python
data_table.plot(x='artist.inverted',y='mean ranking',kind='bar',figsize=(50,50),title='Średni ranking dla utworu')
```

```python
data_mean = data_table.groupby(['genre']).mean()
data_mean
```

Średni ranking dla gatunku

```python
data_mean.plot(y='mean ranking',kind='bar',figsize=(50,50),title='Średni ranking dla gatunku')
```
