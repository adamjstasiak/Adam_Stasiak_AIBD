```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### Zadanie 3


```python
def function(x):
    """
    Definicja funkcji do zadnia 3
    """
    result = []
    for el in x:
        result.append(el**2+5)
    return np.array(result)
```

```python
x1 = np.linspace(-1,1,100)
x2 = np.linspace(-6,6,100)
x3 = np.linspace(0,5,100)

y1 = function(x1)
y2 = function(x2)
y3 = function(x3)
```

```python
f,(ax1,ax2,ax3) = plt.subplots(1,3,figsize=(20,6))
ax1.plot(x1,y1)
ax1.set_title('$x\in(-1,1)$')
ax1.set_xlabel('x')
ax1.set_ylabel('y = f(x)')
ax2.plot(x2,y2)
ax2.set_title('$x\in(-6,6)$')
ax2.set_xlabel('x')
ax2.set_ylabel('y = f(x)')
ax3.plot(x3,y3)
ax3.set_title('$x\in(0,5)$')
ax3.set_xlabel('x')
ax3.set_ylabel('y = f(x)')
f.suptitle('Wykres funkcji $f(x) = x^{2} + 5$')
```
